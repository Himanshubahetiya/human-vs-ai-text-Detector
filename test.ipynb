```python
import pandas as pd 
import pickle
```

```python
with open("text.pkl", "rb")as f:
    data =pickle.load(f)
```

```python
model = data["model"]
tfidf = data["tfidf"]
```

```python
user_text = input("enter the tex")
```

```python
text_vector = tfidf.transform([user_text])
```

```python
model.predict(text_vector)
```
