```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from scipy.sparse import hstack
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
import pickle
```

```python
df = pd.read_csv("AI_Human.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Separate classes
df_human = df[df["generated"] == 0]
df_ai = df[df["generated"] == 1]

# Find minimum count
min_size = min(len(df_human), len(df_ai))

# Sample equal data
df_human_sample = df_human.sample(n=min_size, random_state=42)
df_ai_sample = df_ai.sample(n=min_size, random_state=42)

# Combine
df_balanced = pd.concat([df_human_sample, df_ai_sample])

# Shuffle
df_balance = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Check
print(df_balanced["generated"].value_counts())
```

```python
df = df_balanced
```

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
import numpy as np

def extract_features(text):

    words = text.split()
    
    word_count = len(words)

    sentence_count = text.count(".") + text.count("?") + text.count("!")

    avg_word_len = np.mean([len(w) for w in words]) if words else 0

    avg_sentence_len = word_count / sentence_count if sentence_count > 0 else 0

    unique_ratio = len(set(words)) / word_count if word_count else 0

    punctuation_count = sum(text.count(p) for p in ".,!?;:")

    uppercase_ratio = sum(1 for c in text if c.isupper()) / len(text) if text else 0

    digit_count = sum(c.isdigit() for c in text)

    stopword_ratio = sum(1 for w in words if w.lower() in ENGLISH_STOP_WORDS) / word_count if word_count else 0

    avg_char_sentence = len(text) / sentence_count if sentence_count > 0 else 0

    sentences = text.split(".")
    lengths = [len(s.split()) for s in sentences if len(s.strip()) > 0]
    burstiness = np.std(lengths) if lengths else 0

    repetition = 1 - (len(set(words)) / word_count) if word_count else 0

    return [
        word_count,
        sentence_count,
        avg_word_len,
        avg_sentence_len,
        unique_ratio,
        punctuation_count,
        uppercase_ratio,
        digit_count,
        stopword_ratio,
        avg_char_sentence,
        burstiness,
        repetition
    ]
```

```python
y = df["generated"]
x = df["text"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size= 0.2, random_state= 42, stratify=y
)
```

```python
tfidf = TfidfVectorizer(
    max_features=12000,
    ngram_range=(1,2),
    stop_words="english",
    sublinear_tf=True
)
```

```python
X_train_tfidf = tfidf.fit_transform(x_train)
X_test_tfidf = tfidf.transform(x_test)
```

```python
style_train = np.array(x_train.apply(extract_features).tolist())
style_test = np.array(x_test.apply(extract_features).tolist())
```

```python
scaler = StandardScaler()
style_train_scaler = scaler.fit_transform(style_train)
style_test_scaler = scaler.transform(style_test)
```

```python
X_train = hstack([X_train_tfidf, style_train])
X_test = hstack([X_test_tfidf, style_test])
```

```python
X_train.shape
```

```python
model = LinearSVC(C=1.0, max_iter=5000)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
classification_report(y_test, y_pred)
```

```python
with open ("text.pkl","wb")as f:
    pickle.dump({
       "model" :  model,
        "tfidf" : tfidf,
        "scaler": scaler
    },f)
print("model save successully")
```
